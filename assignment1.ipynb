```python
# Program to calculate total, percentage and grade of a student

# Taking student details
name = input("Enter student name: ")
student_class = input("Enter class: ")

# Taking marks of 5 subjects
sub1 = float(input("Enter marks of Subject 1: "))
sub2 = float(input("Enter marks of Subject 2: "))
sub3 = float(input("Enter marks of Subject 3: "))
sub4 = float(input("Enter marks of Subject 4: "))
sub5 = float(input("Enter marks of Subject 5: "))

# Calculating total and percentage
total = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (total / 500) * 100

# Finding grade using conditional statements
if percentage >= 60:
    grade = 'A'
elif percentage >= 50 and percentage < 60:
    grade = 'B'
elif percentage >= 40 and percentage < 50:
    grade = 'C'
elif percentage >= 33 and percentage < 40:
    grade = 'D'
else:
    grade = 'Fail'

# Displaying result
print("\n----- Student Result -----")
print("Student Name :", name)
print("Class        :", student_class)
print("Total Marks  :", total)
print("Percentage   :", percentage, "%")
print("Grade        :", grade)
```

```python
# Program to concatenate two strings and perform string methods

# Taking two string inputs
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

# Concatenating strings
result = str1 + " " + str2

# Displaying concatenated string
print("\nConcatenated String:", result)

# Performing string methods
print("\n--- String Methods Output ---")

print("lower()      :", result.lower())
print("upper()      :", result.upper())
print("title()      :", result.title())
print("swapcase()   :", result.swapcase())
print("capitalize() :", result.capitalize())
print("casefold()   :", result.casefold())

# center(width)
print("center()     :", result.center(50, "*"))

# count(substring)
print("count('a')   :", result.count('a'))

# endswith()
print("endswith('!'):", result.endswith('!'))

# find(substring)
print("find('a')    :", result.find('a'))

# isalnum()
print("isalnum()    :", result.isalnum())

# isdigit()
print("isdigit()    :", result.isdigit())

# isnumeric()
print("isnumeric()  :", result.isnumeric())

# isspace()
print("isspace()    :", result.isspace())

# replace(old, new)
print("replace()    :", result.replace('a', '@'))
```

```python
# Practice program for Assignment Operators in Python

# Initial value
a = 10
print("Initial value of a =", a)

# += Operator
a += 5
print("After a += 5  :", a)

# -= Operator
a -= 3
print("After a -= 3  :", a)

# *= Operator
a *= 2
print("After a *= 2  :", a)

# /= Operator
a /= 4
print("After a /= 4  :", a)

# %= Operator
a %= 3
print("After a %= 3  :", a)

# **= Operator
a **= 2
print("After a **= 2 :", a)

# //= Operator
a //= 2
print("After a //= 2 :", a)
```
