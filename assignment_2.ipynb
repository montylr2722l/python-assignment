```python
# Dictionary Example

student = {
    "name": "Vishnu",
    "age": 20,
    "course": "Python"
}

print(student)

# Accessing values
print("Name:", student["name"])
print("Age:", student["age"])

# Adding new item
student["city"] = "Jaipur"

print(student)


# Tuple Example

numbers = (10, 20, 30, 40)

print(numbers)

# Accessing elements
print("First Element:", numbers[0])
print("Second Element:", numbers[1])

# Length of tuple
print("Length:", len(numbers))

# Set Example

fruits = {"apple", "banana", "mango", "apple"}

print(fruits)

# Adding element
fruits.add("orange")

print(fruits)

# Removing element
fruits.remove("banana")

print(fruits)
```

```python
# Basic Math Operations

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    return "Cannot divide by zero"

def power(a, b):
    return a ** b

def modulus(a, b):
    if b != 0:
        return a % b
    return "Cannot modulus by zero"

# User input
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

# Operations
print("Addition =", add(a, b))
print("Subtraction =", subtract(a, b))
print("Multiplication =", multiply(a, b))
print("Division =", divide(a, b))
print("Power =", power(a, b))
print("Modulus =", modulus(a, b))
```

```python
#check no. is a palindrome or not   
num = int(input("Enter a number: "))
original_num = num
reverse_num = 0
while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num //= 10
    if original_num == reverse_num:
        print(f"{original_num} is a palindrome.")
    else:
        print(f"{original_num} is not a palindrome.")
        
```
