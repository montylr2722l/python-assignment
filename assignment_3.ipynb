```python
def max_of_three(a, b, c):
    if a >= b and a>= c:
        return a
    elif b >= a and b >= c:
        return b
    else: return c

print(max_of_three(5, 3, 2))
```

```python
def distinct_list(numbers):
    unique_numbers = []
    for num in numbers:
        if num not in unique_numbers:
            unique_numbers.append(num)
    return unique_numbers

print(distinct_list([1, 2, 3, 3, 4, 5, 5]))
```

```python
def multiply_list(numbers):
    result = 1
    for num in numbers:
        result *= num

    return result
   
print(multiply_list([1, 2, 3, 4]))
```

```python
# reverse a string
def reverse_string(s):
    return s[::-1]
print(reverse_string("montylr2722l"))



# check if a number is in a given range
def check_range(num):
    if num in range(1,10):
        return True
    else: return False
print(check_range(5))


# print even numbers in a given range
def print_even(start, end):
    for num in range(start, end + 1):
        if num % 2 == 0:
            print(num)
print_even(1, 10)
```

```python
def print_even_numbers(numbers):
    print("Even numbers are:")
    
    for num in numbers:
        if num % 2 == 0:
            print(num)

# Example usage
my_list = [1, 2, 3, 4, 5, 6, 7, 8]

print_even_numbers(my_list)
```

```python
def check_prime(n):

    if n <= 1:
        print(n, "is not a prime number")
        return

    for i in range(2, n):
        if n % i == 0:
            print(n, "is not a prime number")
            return

    print(n, "is a prime number")


# Example usage
check_prime(7)
check_prime(10)
```

```python
def count_case_letters(text):

    upper = 0
    lower = 0

    for ch in text:

        if ch.isupper():
            upper += 1

        elif ch.islower():
            lower += 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)


# Example usage
count_case_letters("Hello World")
```
