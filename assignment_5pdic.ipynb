```python
# Repeat a tuple three time using *

t = (1, 2, 3)

new_tuple = t * 3

print(new_tuple)
```

```python
# Join three tuples using +

t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)

result = t1 + t2 + t3

print(result)
```

```python
# check weather an element exist in a toupe

t = (10, 20, 30, 40)

element = 30

if element in t:
    print("Element found")
else:
    print("Element not found")
```

```python
# Find total, highest, and lowest without sum(), max(), min()

t = (10, 25, 5, 40, 15)

total = 0
highest = t[0]
lowest = t[0]

for i in t:
    total += i

    if i > highest:
        highest = i

    if i < lowest:
        lowest = i

print("Total =", total)
print("Highest =", highest)
print("Lowest =", lowest)
```

```python
# Filter tuple values greater than 10

n = (3, 14, 7, 22, 9, 41, 18, 5)

result = ()

for i in n:
    if i > 10:
        result += (i,)

print(result)
```

```python
# Count elements in a set without len()

s = {"cat", "dog", "bird", "fish"}

count = 0

for i in s:
    count += 1

print("Total elements =", count)
```

```python
# Combine two sets (Union)

s1 = {1,2,3}
s2 = {4,5,6}

result = s1|s2

print(result)
```

```python
# Find common elemets in two sets (intersection)

s1 = {1,2,3,4}
s2 = {3,4,5,6}
 
result = s1 & s2
print(result)
```

```python
# Find elements that are in either set but not both (Symmetic difference)

s1 = {1,2,3,4}
s2 = {3,4,5,6}

result = s1^s2
print(result)
```
