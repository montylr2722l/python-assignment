```python
import pandas as pd

#pandas series form dictonary
data = {
    "name": "Alice",
    "age": 30,  
    "city": "New York"
}
series = pd.Series(data)
print(series)

#pandas series form list
number = [10, 20, 30, 40, 50]
series = pd.Series(number)
print(series)

print (series[0])  # Accessing the first element of the series
print (series[1]) #output: 20
print(series[1:3])  # Accessing a slice of the series (20 and 30)

```

```python
#DataFrame from list 
data = [
    ["Alice", 30, "New York"],
    ["Bob", 25, "Los Angeles"],
    ["Charlie", 35, "Chicago"]
]
df = pd.DataFrame(data, columns=["name", "age", "city"])
print(df)

#DataFrame from dictonary
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [30, 25, 35],
    "city": ["New York", "Los Angeles", "Chicago"]
}
df = pd.DataFrame(data)
print(df)

#DataFrame from list of dictonary
data = [    {"name": "Alice", "age": 30, "city": "New York"},
    {"name": "Bob", "age": 25, "city": "Los Angeles"},
    {"name": "Charlie", "age": 35, "city": "Chicago"}           
]
df = pd.DataFrame(data)
print(df)

#DataFrame from list of list
data = [  ["Alice", 30, "New York"],
    ["Bob", 25, "Los Angeles"], 
    ["Charlie", 35, "Chicago"]      
]   
df = pd.DataFrame(data, columns=["name", "age", "city"])
print(df)

#DataFrame from list of tuple
data = [    ("Alice", 30, "New York"),  
    ("Bob", 25, "Los Angeles"), 
    ("Charlie", 35, "Chicago")      
]
df = pd.DataFrame(data, columns=["name", "age", "city"])
print(df)

#Selecting rows based on condition
result = df[df["age"] > 30]
print(result)   

#selecting any row using iloc[]
result = df.iloc[1]  # Accessing the second row (index 1)
print(result)

#limited row selection with given columns
result = df.loc[0:1, ["name", "city"]]  # Accessing the first two rows and specific columns
print(result)

#drop row based on condition
result = df.drop(df[df["age"] < 30].index)  # Dropping rows where age is less than 30
print(result)

#insert row at given position
new_row = {"name": "David", "age": 28, "city": "San Francisco"}
df = pd.concat([df.iloc[:1], pd.DataFrame([new_row]), df.iloc[1:]]).reset_index(drop=True)  # Inserting the new row at index 1
print(df)

#create a list from rows in pandas dataframe
row_list = df.values.tolist()  # Convert DataFrame rows to a list of lists
print(row_list)
```

```python
#Data Iteration
import pandas as pd
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [30, 25, 35],
    "city": ["New York", "Los Angeles", "Chicago"]
}
#method 1: iterrows()
df = pd.DataFrame(data)
for index, row in df.iterrows():
    print(f"Index: {index}, Name: {row['name']}, Age: {row['age']}, City: {row['city']}")

#method 2: itertuples()
for row in df.itertuples():
    print(f"Index: {row.Index}, Name: {row.name}, Age: {row.age}, City: {row.city}")    
print(df)
```
