```python
df1 = pd.DataFrame({
    'Id': [1, 2, 3, 4, 5],
    'Name': ['vishnu', 'piya', 'aman', 'rishi', 'divya']
})

df2 = pd.DataFrame({
    'Id': [2, 3, 4, 5, 6],
    'city': ['delhi', 'mumbai', 'kolkata', 'chennai', 'bangalore']
})

print   ("Dataframe 1:")
print(df1)
print("Dataframe 2:")
print(df2)

inner_merge = pd .merge(df1, df2, on='Id', how= 'inner')
print("Inner Merge:")
print(inner_merge)
left_merge = pd.merge(df1, df2, on='Id', how='left')
print("Left Merge:")    
print(left_merge)


#indexd based join
df1_index = df1.set_index('Id')
df2_index = df2.set_index('Id')

print   ("Dataframe 1 with Id as index:")
print(df1_index)
print("Dataframe 2 with Id as index:")
print(df2_index)

joined = df1_index.join(df2_index)
print   ("Joined Dataframe:")
print(joined)
```

```python
df1 = pd.DataFrame({
    'Id': [1,2],
    'Name': ['vishnu', 'piya']
})

df2 = pd.DataFrame({
    'Id': [3,4],
    'Name':['aman', 'rishi']
})

df3 = pd.DataFrame({
    'Id': [1,2,3,4],
    'marks': [39,43,35,76] 
})

combined_df = pd.concat([df1, df2])
print("Combined DataFrame:")    
print(combined_df)
final_df = pd.merge (combined_df, df3, on='Id')
print("Final DataFrame after merging with marks:")
print(final_df)
```
