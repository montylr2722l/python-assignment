```python
# > replace Nan with 0 and Interchange 3 row and 3 column of 2D arrary [[6,-8,73,-110], [np.nan, -8, 0, 94]]
import numpy as np
arr = np.array([[6, -8, 73, -110],[np.nan, -8, 0,  94]])
arr = np.nan_to_num(arr, nan= 0)
print(arr)
arr = arr.T
print(arr)
```

```python

# 3) Replace NaN values with average of columns 
arr1=np.array([[0,1,np.nan,3],[np.nan,5,6,7]])
print(arr1)
col_mean = np.nanmean(arr1, axis=0)
inds = np.where(np.isnan(arr1))

arr1[inds] = np.take(col_mean, inds[1])

print(arr1)
```

```python

# 5) Study the following program 
# import numpy as np 
# # create a numpy 1d-arrays 
# arr1 = np.array([3, 4]) 
# arr2 = np.array([1, 0]) 
# # find average of NumPy arrays
# avg = (arr1 + arr2) / 2 
# print("Average of NumPy arrays:\n", avg)
# -> Calculate average mean median mode values of two NumPy 2d-arrays
from scipy import stats 
arr3=np.array([[1,2,3,4],[5,6,7,8]])
arr4=np.array([[11,12,13,14],[15,16,17,18]])
print(arr3)
print(arr4)
avg=(arr3+arr4)/2
print(avg)

combined=np.concatenate((arr3.flatten(),arr4.flatten()))
print("Mean:",np.mean(combined))
print("Median:",np.median(combined))
print("Mode:",stats.mode(combined))
```

```python
# 6) Solve the following equation using linalg() and inverse Matrix method 
# x - 2y + 3z = 9 
# -x + 3y - z = -6 
# 2x - 5y + 5z = 17
import numpy as np

A = np.array([
    [1,-2,3],
    [-1,3,-1],
    [2,-5,5]
])

B = np.array([
    [9],
    [-6],
    [17]
])

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

print("Solution:")
print(X)
```
