```python
n1 = int(input("enter a number:"))
operator = input("enter an operator: ")
n2 = int(input("enter another number: "))
if operator == "+":
    print(n1 + n2)
elif operator == "-":
    print(n1 - n2)  
elif operator == "*":
    print(n1 * n2)  
elif operator == "/":
    print(n1 / n2) 
elif    operator == "%":
    print(n1 % n2)
else:
    print("invalid operator")
```
