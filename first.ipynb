```python
# dictionary
d = {"name": "john", "age": 25, "city": "new york"}
print(d["name"])
print(d.get("age"))
print(d.get("city"))
print(d.get("country", "not found")) # it will return "not found" because the key "country" is not present in the dictionary        
print(d.keys()) # it will print all the keys of the dictionary
print(d.values()) # it will print all the values of the dictionary
print(d.items()) # it will print all the key-value pairs of the dictionary
print(d.pop("age")) # it will remove the key "age" and return its value
print(d) # it will print the dictionary after removing the key "age"
print(d.popitem()) # it will remove the last key-value pair from the dictionary and return it as a tuple
print(d) # it will print the dictionary after removing the last key-value pair
print(d.clear()) # it will remove all the key-value pairs from the dictionary
print(d) # it will print the empty dictionary
print(d.setdefault("name", "john")) # it will return the value of the key "name" if it is present in the dictionary, otherwise it will insert the key "name" with the value "john" and return "john"
print(d) # it will print the dictionary after setting the default value for the key "name"
print(d.update({"age": 25, "city": "new york"})) # it will update the dictionary with the key-value pairs from the given dictionary
print(d) # it will print the updated dictionary
```

```python
#Tuple

#Tuples are immutable, ordered collection of elements. They are defined using parentheses () and elements are separated by commas. Tuples can contain elements of different data types.
t1 = (1,2,3,4,5)
t2 = ("vishnu", 43, 3.14, True, [1,2,3], (4,5,6), {"name": "john", "age": 25})
print(t1, type(t1))
print(t2, type(t2))
for i in t1:
    print(i)
for i in t2:
    print(i)
count = 0
for i in t2:
    count += 1
print(count)
print(t1 + t2) # it will concatenate the two tuples
print(t1 * 2) # it will repeat the tuple t1 two times
print(t1[::-1]) # it will print the tuple in reverse order
print(t1[0]) # it will print the first element of the tuple
# del t1 # it will delete the tuple t1
# print(t1) # it will raise an name error because t1 is deleted
print(list(t1)) # it will convert the tuple t1 to a list
print(tuple(list(t1))) # it will convert the list back to a tuple   
print(tuple("hello")) # it will convert the string "hello" to a tuple of characters
print(min(t1)) # it will print the minimum element in the tuple t1
print(max(t1)) # it will print the maximum element in the tuple t1
print(t1.index(3)) # it will print the index of the first occurrence of 3 in the tuple t1
print(t1.count(2)) # it will count the number of occurrences of 2 in the tuple t1
print(sum(t1)) # it will print the sum of all the elements in the tuple t1
```

```python
# Set
s1 = {1,2,3,4,5}
s2 = {"vishnu", 43, 3.14, True, (4,5,6), frozenset({1,2,3}), frozenset({"name": "john", "age": 25})}
s3 = set({}) # it will create an empty set
print(s1, type(s1))
print(s2, type(s2))
print (s1.union(s2)) # it will return a new set that contains all the elements from both sets
print(s1.intersection(s2)) # it will return a new set that contains only the elements that are present in both sets
print(s1.difference(s2)) # it will return a new set that contains only the elements that are present in s1 but not in s2
for i in range(7, 10):
    s3  .add(i) # it will add the element i to the set s3   
print(s3)
s1.clear() # it will remove all the elements from the set s1
print(s1) # it will print the empty set
```

```python
import csv

# file handling
file = open("file.txt", "w")  
# file.txt create hogi write mode me

file.write("this is a file handling example")  
# file ke andar text write hoga

file.close()  
# file close karna important hai


# file read karna
with open("file.txt", "r") as file:
    data = file.readlines()

    for line in data:
        word = line.split()
        print(word)


# CSV file read karna
with open("data.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
# Exception handling
try:
    n1 = int(input("enter a number: "))
    result = 10 / n1
    print(result)
except ValueError:
    print("Please enter a valid integer.")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
else:
    print("The operation was successful.")
```
