```python

# > Create a 4×4×3 array and extract value at its second set, first row and last column
arr4 = np.arange(48).reshape(4,4,3)
print(arr4)
print(arr4[1,0,-1])
```

```python

# > Create a 4×4 and Extract Odd Rows and Even Columns 
arr5 = np.arange(1,17).reshape(4,4)
print("\n4x4 Array:")
print(arr5)
print(arr5[0::2,1::2])
```

```python
# slice the first two rows and first two columns of second set from a 4*4*3 array
arr6 = arr4[1,0:2,0:2]
print(arr6)
```

```python
# replace all odd numbers in a numpy array with -1 by itterating usign for loop [[23,56,78,93], [71,82,13,24]]
arr7 = np.array([[23,56,78,93],[71,82,13,24]])
print(arr7)
for i in range (arr7.shape[0]):
    for j in range(arr7.shape[1]):
     if arr7[i, j] % 2 != 0:
      arr7[i, j] = -1
print(arr7)
```

```python
# > get the indices of non-zero elements in an array [1,0,2,0,3,0,4]
arr8 = np.array([1,0,2,0,3,0,4])
indices = np.nonzero(arr8)
print(indices)
```
