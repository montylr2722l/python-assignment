```python
#Dataframe

import pandas as pd
student = {
    "name" : ["Alice", "Bob", "Charlie"],
    "age" : [20, 21, 22],
    "city" :["New york", "Los Angeles", "Chicago"]
}

df = pd.DataFrame(student)
print(df)
print(df.shape)
print(df[df.columns[2]])
```

```python
import pandas as pd
df1 = pd.DataFrame({'A': [1,2,3], 'B': [4,5,6]})
df2 = pd.DataFrame({'A': [7,8,9], 'B': [10,11,12]}, index = [0,1,2])
print(df1)
print(df2)
print("\n Addition \n "  , df1+df2)
```

```python
#pivot table
import pandas as pd 
data = {
    "name" : ["vishnu", "piya" , "aman", "khushal" ,"ravi"],
    "age" :[ 20, 21, 22, 23, 24],
    "city" : ["delhi", "mumbai", "kolkata", "chennai", "bangalore"],
    "course" : ["python", "java", "c++", "javascript", "ruby"]
}
df = pd.DataFrame(data)
print(df)
pivot_table = df.pivot_table(index = "city", values = "age", aggfunc = "mean")
print(pivot_table)

```
